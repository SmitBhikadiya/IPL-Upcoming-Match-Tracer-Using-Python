## IPL Schedule ##

```python
import requests # to get html page content from the url
from bs4 import BeautifulSoup as bs
import pandas as pd
import datetime
from datetime import date
```

```python
ipl_url = "https://www.iplt20.com/matches/schedule/men"
page_obj = requests.get(ipl_url)
```

```python
html = page_obj.content
```

```python
soup = bs(html,'html.parser')
print(soup)
```

```python
li = []
#length_of_team = len(team_)
cur_time = datetime.datetime.now()
for match_list in soup.find_all(class_="match-list__item"):
    team_ = match_list.find_all(class_="fixture__team-name")
    match_no = match_list.find(class_="fixture__description").getText()
    match_info = match_list.find(class_="fixture__info")
    match_loc = match_info.span.getText()[8:]
    match_time = match_info.strong.getText()[:5]
    length_of_team = len(team_)
    team = []
    for x in range(0,(length_of_team//2)+1,2):
        if (length_of_team%2)!=0 and (length_of_team//2)==x:
            tname = team_[x].getText()
        else:
            tname = team_[x].getText()
            abbr_name = team_[x+1].getText()
            if x==2:
                team.append(f'{cur_time.day}-{cur_time.month}-{cur_time.year}')
                team.append(f'{match_no}')
                team.append(f'{temp} vs {tname}')
                team.append(f'{atemp} vs {abbr_name}')
                team.append(f'{str(int(match_time[:2])-12)}{match_time[2:]}')
                team.append(f'{match_loc}')
            else:
                atemp = abbr_name
                temp = tname
    if match_time != "15:30":
        cur_time += datetime.timedelta(days=1)
    li.append(team)

for team in li:
    print(str(team).lower())
```

```python
df = pd.DataFrame(li)
df.columns = ["matchDate","matchNo","matchVS","abbrName","matchTime","matchLoc"]
df.set_index("matchDate")
```

## remaining matches count ##

```python
# intialization
team_series = list(df.loc[:,"abbrName"])
team_cnt = {}
# team_series = ["rr","DC","rr","DC","RCB","rr"] # for testing purpose

# logic
for match in team_series:
    teams = match.split(" vs ")
    for team in teams:
        if team in team_cnt:
            team_cnt[team] += 1  
        else:
            team_cnt[team] = 1
            
# output 
team_cnt
```
